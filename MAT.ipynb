```python
#Loading Image
import cv2
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
```

```python
# Path to your image
path = "/Users/abhishek/Desktop/CS_Assignment/Images/Flower_binary.png"   # replace with your file path

# Read the image
img = mpimg.imread(path)

# Display the image in grayscale
plt.imshow(img, cmap="gray")
plt.axis("off")
plt.show()
```

```python
#Resize Image
resized_image = cv2.resize(img, (224, 224))
plt.imshow(resized_image, cmap="gray")
plt.axis("off")
plt.show()
```

```python
def manhattan_distance(resized_image, i, j):
    rows, cols = resized_image.shape
    min_dist = float('inf')
    
    for x in range(rows):
        for y in range(cols):
            if resized_image[x][y] == 0:  # background pixel
                dist = abs(i - x) + abs(j - y)  # Manhattan distance
                # dist = max(abs(i - x), abs(j - y)) for chessboard distance
                if dist < min_dist:
                    min_dist = dist
    
    return min_dist
```

```python
# Manhattan Distance Calculation
for i in range(224):
    for j in range(224):
        if resized_image[i][j] == 1:
            resized_image[i][j] = manhattan_distance(resized_image, i, j)
        else:
            resized_image[i][j] = 0
            
```

```python
def maximum_filering(img):
    img2=img.copy()
    for i in range(1,223): #Original size 224x224
        for j in range(1,223):
            arr=[img[i-1][j-1],img[i-1][j],img[i-1][j+1],
                 img[i][j-1],img[i][j],img[i][j+1],
                 img[i+1][j-1],img[i+1][j],img[i+1][j+1]]
            img2[i][j]=max(arr)
    return img2
```

```python
from scipy.ndimage import maximum_filter

def medial_axis_transform(dist_map):
    # neighborhood maximum
    local_max = (dist_map == maximum_filering(dist_map))
    # keep only non-zero values
    return local_max * dist_map
```

```python
mat = medial_axis_transform(resized_image)

plt.subplot(1,2,1)
plt.title("Distance Transform")
plt.imshow(resized_image, cmap="hot")
plt.colorbar()

plt.subplot(1,2,2)
plt.title("MAT")
plt.imshow(mat, cmap="hot")
plt.colorbar()

plt.show()
```
